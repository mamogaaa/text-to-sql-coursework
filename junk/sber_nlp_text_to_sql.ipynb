```python
import pandas as pd
import numpy as np
import json
from templates import get_data
from query_generator import generate_query_samples_for_intent

data = get_data()

rows = []
for intent in data:
    rows = [*rows, *generate_query_samples_for_intent(intent)]

df = pd.DataFrame(rows)
df
```

```python
# generate NER train data
!mkdir -p output/ner_data
from sklearn.model_selection import train_test_split

ner_data = list(df['train_data'].sample(300, random_state=42))
ner_train, ner_test = train_test_split(ner_data, test_size=0.2, random_state=42)
ner_test, ner_val = train_test_split(ner_test, test_size=0.5, random_state=42)

ner_train_data = "\n\n".join(ner_train)
ner_test_data = "\n\n".join(ner_test)
ner_val_data = "\n\n".join(ner_val)

with open('output/ner_data/train.txt', 'w') as f:
    f.write(ner_train_data)
with open('output/ner_data/test.txt', 'w') as f:
    f.write(ner_test_data)
with open('output/ner_data/valid.txt', 'w') as f:
    f.write(ner_val_data)
```

```python
from sklearn.model_selection import train_test_split

X, y = df[['query']], df[['intent']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(text_features=['query'], loss_function='MultiClass', eval_metric='MultiClass')
clf.fit(X_train, y_train, eval_set=(X, y))
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
res = X_test.copy()
res['intent_pred'] = y_pred
res['intent'] = y_test
res
```
